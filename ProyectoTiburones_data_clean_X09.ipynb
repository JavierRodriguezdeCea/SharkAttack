```python
# Quito los espacios antes y después
tabla['activity'] = tabla['activity'].str.strip()

# Cuento cuántas entradas son nulas o vacías en 'activity'
entradas_invalidas = tabla['activity'].isna().sum() + (tabla['activity'] == '').sum()
print(f"Número de entradas inválidas {entradas_invalidas}")

# Defino una función para agrupar deportes según la actividad. Importante uso ANY para que lea los valores y si tiene esa palabra lo asigne a un tipo de deporte
def categorize_sport(activity):
    if pd.isna(activity) or activity == '':
        return 'Unknown'
    activity = activity.lower()
    if any(word in activity for word in ['surf', 'swimming', 'diving', 'snorkel', 'paddle', 'kayak']):
        return 'Water Sports'
    elif any(word in activity for word in ['fishing', 'spearfishing']):
        return 'Fishing'
    elif any(word in activity for word in ['hiking', 'walking', 'running', 'jogging']):
        return 'Land Sports'
    elif any(word in activity for word in ['air', 'paragliding']):
        return 'Aerial Sports'
    else:
        return 'Other'

# Aplico la función para crear la columna 'sport_type'
tabla['sport_type'] = tabla['activity'].apply(categorize_sport)

tabla[['activity', 'sport_type']].head()
```

```python
# Función para extraer solo la raza de tiburón
def extract_shark_species(species):
    if pd.isna(species) or species == '':
        return 'unknown'

    # Pasamos a minúsculas para simplificar el proceso
    species = species.lower()

    # Buscamos patrones comunes de razas de tiburón
    if 'white' in species:
        return 'white shark'
    elif 'tiger' in species:
        return 'tiger shark'
    elif 'bull' in species:
        return 'bull shark'
    elif 'reef' in species:
        return 'reef shark'
    elif 'hammerhead' in species:
        return 'hammerhead shark'
    elif 'nurse' in species:
        return 'nurse shark'
    elif 'mako' in species:
        return 'mako shark'
    elif 'whale' in species:
        return 'whale shark'
    elif 'lemon' in species:
        return 'lemon shark'
    else:
        # Si no se encuentra una raza específica, retornar 'unknown'
        return 'unknown'

# Aplicamos la función para simplificar la columna 'species'
tabla['species_simplified'] = tabla['species'].apply(extract_shark_species)

# Verificamos los resultados
print(tabla[['species', 'species_simplified']].head())
```

```python
# Limpiamos la columna 'fatal' simplificando los resultados y guardando la limpieza en una columna nueva para no modificar los datos
def limpiar_fatal(value):
    # Eliminamos espacios en blanco al principio y al final
    value = str(value).strip().upper()

    # Revisamos si el valor es 'Y' o 'N' y normalizarlo
    if value in ['Y', 'Y X 2']:
        return 'Y'
    elif value in ['N', ' N', 'NQ', 'Nq', 'n']:
        return 'N'
    elif value in ['UNKNOWN', 'M', 'F', '2017']:
        return 'UNKNOWN'
    else:
        return 'UNKNOWN'

# Aplicar la función de limpieza a la columna 'fatal'
tabla['fatal_cleaned'] = tabla['fatal'].apply(limpiar_fatal)

tabla['fatal_cleaned'].value_counts()
```

```python
def clasificar_hora(value):
    if value == "":
        return "unnkown"

    value = int(value)

    # Revisamos si el valor es se encuentra en unos rangos para determinar si es mañana, tarde, noche o desconocido
    if (((value >= 0) and (value < 6))) or ((value >= 22) and (value < 24)):
        return "night"

    elif ((value >= 6) and (value < 14)):

        return "morning"

    elif ((value >= 14) and (value < 22)):

        return "afternoon"

    else:

        return "unnkown"

# Aplicar la función de limpieza a la columna 'fatal'
tabla['time_definitive'] = tabla['time_cleaned'].apply(clasificar_hora)

tabla['time_definitive'].unique()
```
